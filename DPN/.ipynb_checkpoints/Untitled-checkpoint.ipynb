```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```

```python
def binarizasia (img): #на вход принимаем серое изображение
    if img.max() > 1:
        color = 1
    else:
        color = 255
    line_image = np.array(img).flatten()#переводим матрицу в строку для того, чтобы вычислить
    list_hist = np.histogram(line_image - min(line_image), int((max(line_image)-min(line_image))*color))# гистограмму значений
                                                            #на которые отличаются писксени от наименьшего
    list_count = list_hist[0] #количество пикселей, попавших в интервал
    list_values = list_hist[1][1:]#правая граница интервалов (левая в данном случае всегда 0)

    mean1 = sum(list_count*list_values) 
    mean2 = sum(list_count)

    alpha = 0
    beta = 0
    maxSigma = 0
    threshold = 0
    for i in range(len(list_count)): #находим порог у гистограмы

        alpha = alpha + list_values[i]*list_count[i]
        beta = beta + list_count[i]

        w1 = beta / mean2
        a = alpha / beta - (mean1 - alpha) / (mean2 - beta);

        sigma=w1*(1-w1)*a*a;#подсчет междуклассовой дисперсии

        if(sigma>maxSigma):
            maxSigma = sigma;
            threshold = list_values[i];

    threshold = threshold + min(line_image) #и порог у изображения
    
    
    img[img < threshold] = 0#производим бинаризацию изображения
    img[img > threshold] = 1
    
    return img
```
